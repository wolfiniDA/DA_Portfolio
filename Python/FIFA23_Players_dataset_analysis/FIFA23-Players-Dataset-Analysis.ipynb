## Number of all players by their overall ratings

### Player's distribution by their overall (Scatterplot)

## Best players by overall rating

```python
best_pl= df[df['Overall'] == 91]
best_pl
```

### Best player's value (Bar chart)

```python
ax= sns.barplot(data=best_pl, x='Known As', y='Value(in Euro)')
ax.set(title="Best Player's Value")
plt.xticks(fontsize='8')

def format_func(value, tick_number):
    if value >= 1000000:
        value = int(value/1000000)
        return f'{value}M'
    elif value >= 1000:
        value = int(value/1000)
        return f'{value}K'
    else:
        return value

ax.yaxis.set_major_formatter(ticker.FuncFormatter(format_func))
```

### Best player's position (Pie chart)

## Players with the highest and the lowest potential

```python
def growth(x, y):
    return x - y


df['Potential Growth'] = df.apply(lambda row: growth(row['Potential'], row['Overall']), axis=1)

df[['Known As', 'Full Name', 'Overall', 'Potential', 'Potential Growth']].sort_values(by='Potential Growth', ascending=False)
```

## Best players by ratio beetween their value and overall rating

```python
def overall_per_price(x, y):
    return x / y


df = df[df['Value(in Euro)'] != 0]

df['Overall Per Price'] = df.apply(lambda row: overall_per_price(row['Value(in Euro)'], row['Overall']), axis=1)

best_ratio = df[['Known As', 'Full Name', 'Club Name', 'Overall', 'Value(in Euro)', 'Overall Per Price']].sort_values(by='Overall Per Price').head(10)
best_ratio
```

## Worst players by ratio beetween their value and overall rating

```python
worst_ratio = df[['Known As', 'Full Name', 'Club Name', 'Overall', 'Value(in Euro)', 'Overall Per Price']].sort_values(by='Overall Per Price',ascending=False).head(10)
worst_ratio
```
