# Nuclear Explosions Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\projects\\nuclear_explosions.csv")
```

Data.Source	-	Source that reported the explosion event<br>
WEAPON SOURCE COUNTRY	-	Country deploying the nuclear device<br>
WEAPON DEPLOYMENT LOCATION	-	Region where the nuclear device was deployed<br>
Location.Coordinates.Latitude	-	Latitude position<br>
Location.Coordinates.Longitude	-	Longitude position<br>
Data.Magnitude.Body	-	Body wave magnitude of explosion (mb)<br>
Data.Magnitude.Surface	-	Surface wave magnitude of explosion (Ms)<br>
Location.Coordinates.Depth	-	Depth at detonation in Km (could be underground or above ground) -- positive = depth (below ground), negative = height (above ground)<br>
Data.Yield.Upper	-	Explosion yield upper estimate in kilotons of TNT<br>
Data.Purpose	-	Purpose of detonation: COMBAT, FMS, ME, PNE, SAM, SSE, TRANSP, WE, WR<br>
Data.Name	-	Name of event or bomb<br>
Data.Type	-	Type - method of deployment<br>
Date.Day	-	Day of detonation<br>
Date.Month	-	Month of detonation<br>
Date.Year	-	Year of detonation<br>

## Check str 'Nan' values in columns

```python
search_text = 'Nan'

contains_search = df.applymap(lambda cell: search_text == str(cell))

rows_with_search = contains_search.any(axis=1)

result_df = df[rows_with_search]

result_df
```

## Str 'Nan' values count

```python
count_per_column = df.applymap(lambda cell: search_text == str(cell)).sum()

count_per_column
```

## Count of all nuclear explosions by country

```python
count_country = df.groupby('WEAPON SOURCE COUNTRY')['WEAPON SOURCE COUNTRY'].count().sort_values(ascending=False)
count_country
```

### Barchart

```python
bars = plt.bar(count_country.index, count_country, color=['Red', 'Orange','Yellow', 'YellowGreen', 'Green', 'Blue', 'Violet'])

plt.xlabel('Country')
plt.ylabel('Count')
plt.title('All nuclear explosions by country')
plt.bar_label(bars, labels=count_country)

plt.show()
```

## Count of all nuclear explosions by location

```python
count_location = df.groupby('WEAPON DEPLOYMENT LOCATION')['WEAPON DEPLOYMENT LOCATION'].count().sort_values(ascending=False)
count_location
```

## Count of all nuclear explosions by year

```python
count_year = df.groupby('Date.Year')['Date.Year'].count().sort_values(ascending=False)
count_year
```

### Barchart

```python
plt.figure(figsize=(10,5),dpi=200)
plt.xticks(rotation=90)
sns.barplot(x = count_year.index,y = count_year,palette='Set1')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('All nuclear explosions by year')
```

## Count of all nuclear explosions by method of deployment

```python
count_method = df.groupby('Data.Type')['Data.Type'].count().sort_values(ascending=False)
count_method
```

### Barchart

```python
plt.bar(count_method.index, count_method)

for i, v in enumerate(count_method):
    plt.text(i, v + 1, str(v), ha='center', va='bottom')

plt.xlabel('Method of explosion')
plt.ylabel('Count')
plt.title('All nuclear explosions by method')
plt.xticks(rotation=90)

plt.show()
```

## Count of all nuclear explosions by purpose of detonation

```python
count_purpose = df.groupby('Data.Purpose')['Data.Purpose'].count().sort_values(ascending=False)
count_purpose
```

### Top 8

```python
count_purpose_top8 = count_purpose.head(8)
count_purpose_top8
```

### Barchart

```python
plt.figure(figsize=(10,5),dpi=100)
plt.xticks(rotation=45)
sns.barplot(x = count_purpose_top8.index,y = count_purpose_top8,palette='Set2')
plt.xlabel('Purpose of detonation')
plt.ylabel('Count')
plt.title('Top 8 nuclear explosions by purpose of detonation')
for i, v in enumerate(count_purpose_top8):
    plt.text(i, v + 1, str(v), ha='center', va='bottom')
```
